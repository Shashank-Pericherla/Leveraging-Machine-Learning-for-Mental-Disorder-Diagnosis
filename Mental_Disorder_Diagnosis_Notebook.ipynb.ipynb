There is an unecessary column present in our dataset. Along with it, education and IQ feature columns have missing values.

### Handling Missing values:

We have to remove the uneccesary column along with the eeg date feature as there is no use with the column. Then we fill the missing values in the IQ and education columns with the mean values as these columns consist of discrete numerical values. 

```python
#dropping the uneccesary columns 
data = data.drop(['Unnamed: 122','eeg.date'],axis=1)   

#filling the missing values with the mean value of the column                  
data['IQ'] = data['IQ'].fillna(data['IQ'].mean())                          
data['education'] = data['education'].fillna(data['education'].mean())
```

# Data Preprocessing

The classes in our dataset are in string format and machine learning models are trained on either int or float numbers. So, we encode the classes into numbers.

```python
from sklearn.preprocessing import OrdinalEncoder            #importing the encoder from sklearn

encoder = OrdinalEncoder()                                  #initializing an instance of the encoder  

encoder.fit(data[['sex']])                                  #fitting the encoder with our data
data['sex']= encoder.transform(data[['sex']])               #encoding the column containing sex

encoder.fit(data[['main.disorder']])
data['main.disorder']= encoder.transform(data[['main.disorder']])         #encoding the column containing disorder
```

Splitting the dataset into training features and labels.

```python
X = data.drop(['main.disorder'],axis=1)    #dropping the disorder classes returns the features
Y = data['main.disorder']                  #getting the classes 

X.shape, Y.shape                           #checking the shape of the data
```

Now to tackle the class imbalance we use sinthetic minority oversampling technique(SMOTE) to oversample the minority classes.

### Oversampling minority classes:

For this task we are implimenting a unique method of smote called SMOTEEN which is a combination of oversampling the data and undersampling the data by removing the outliers.

```python
from imblearn.combine import SMOTEENN                              #importing the oversampler for imblearn library 

smote = SMOTEENN(sampling_strategy='all', random_state= 42)        #initializing the oversampler with the necessary hyper parameters 
X_resampled , Y_resampled = smote.fit_resample(X,Y)                #resampling the data

X_resampled.shape, Y_resampled.shape                               #checking the shape of the resampled data
```

```python
Y_resampled = Y_resampled.astype(np.int64)   #changing the dtype from float32 to int64 for faster computing
```

```python
#Visualizing the class distribution in our dataset after oversampling
resampled_class_counts = Y_resampled.value_counts().sort_index()

plt.figure(figsize=(8,6),dpi=100)
plt.bar(x=resampled_class_counts.index,height= resampled_class_counts.values,)
plt.title("Distribution of the classes of disorders after oversampling")
plt.xlabel("Disorders")
plt.ylabel("Count")
```

As we can see, the class imbalance has been eliminated.

### Feature Scaling:

For this task we make use Robust scaler which is a variation of standard scaler and uses inter quatile range to calculate the scaling factor. This method is more robust and immune to outliers.

```python
from sklearn.preprocessing import RobustScaler                       #importing the scaler from sklearn

scaler = RobustScaler()                                              #initializing the scaler 
col_names = X_resampled.columns                                      #storing the column names for future use
X_resampled = scaler.fit_transform(X_resampled)                      #applying the scaler to the data
X_resampled = pd.DataFrame(X_resampled, columns=col_names)           #converting the data to dataframe
```

### Feature Selection (using Elastic Net):

Let's visualize the correlation between the features.

The initial columns contains categorical variables whereas the later contain numerical variables. So we have to use different correlation metrics for both.

```python
data.iloc[:,:5].corr(method='spearman') 
```

```python
data.iloc[:,5:].corr(method='pearson')
```

Elastic net combines the advantages of lasso and ridge regularization to handle and assign the right coefficient for each features. we can control the trade-off between the two regularization methods in elastic net.

```python
# importing the elastic net module 
from sklearn.linear_model import ElasticNet

elastic_net = ElasticNet( alpha =0.01, l1_ratio= 0.001)                        #initializing elastic net with the appropriate l1 and l2 ratios
elastic_net.fit(X_resampled, Y_resampled)                                      #fitting the elastic net model on the data
coef_abs = np.abs(elastic_net.coef_)                                           #calculating the coefficients for each features
sorted_coef_abs = np.argsort(coef_abs)[::-1]                                   #sorting the coefficients in descending order
top_500_features = sorted_coef_abs[:500]                                       #selecting the top 500 features
top_500_feature_names = [X_resampled.columns[i] for i in top_500_features]     #creating a list of the feature names
df=[]                                                                          #creating a list to store the new top 500 features 

for j in list(top_500_feature_names):                                          #loop to append all the top 500 features in a new dataframe
   df.append(X_resampled[j])
df = pd.concat(df, axis = 1)
```

```python
top_500_feature_names
```

### Data Splitting:

```python
from sklearn.model_selection import train_test_split

#splitting the data into training and testing sets in the ratio of 80:20
X_train , X_test, y_train, y_test = train_test_split(df, Y_resampled , test_size = 0.2, random_state = 0)   
```

# Machine Learning:

### Model Selection:

Since our dataset is structured and small, machine learning models are preferred over deep learning. And since we have a large number of training features, tree based models are suitable and ensemble models like gradient boosting models are preferred.  

LightGBM model works well with large number of features and training the model is lighter, faster, and much efficient compared to other boosting models like XGBoost or CatBoost.

```python
#importing the models
import xgboost as xgb
import lightgbm as lgm
                                  
from sklearn.model_selection import cross_val_score,RepeatedKFold                         #importing the cross validation method
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report       #importing the evaluation metrics

RepeatedKFold = RepeatedKFold(n_splits=5, n_repeats=10, random_state = 42)                #intializing an instance of the repeated K fold with the necessary hyper parameters 
```

### LightGBM

```python
model = lgm.LGBMClassifier(device_type='gpu',       #leveraging GPU acceleration for faster training
                       n_estimators=100,            #number of trees in the forest
                       max_depth=-1,                #maximum depth of the tree
                       learning_rate=0.1,           #learning rate
                       reg_alpha=0.1,               #L1 regularization penalty to restrict the model from overfitting
                       reg_lambda=1,                #L2 regularization penalty to restrict the model from overfitting
                       n_jobs=-1,                   #number of jobs to run in parallel
                       metric='multi_logloss',      #metric to optimize
                       )
```

```python
model.fit(X_train, y_train,eval_set=[(X_train,y_train),(X_test, y_test)]) 
```

Now we check the performance of our model on unseen data.

```python
predictions = model.predict(X_test)                         
print('Accuracy Score:',accuracy_score(y_test, predictions))
report = classification_report(y_test, predictions) 
print(report)
```

Great! we got an accuracy of 96.5 with great precision and recall scores.

```python
#Visualizing the training and validation loss
lgm.plot_metric(model,figsize=(10,6))
```

```python
#Visualizing a random tree from boosting model 
lgm.create_tree_digraph(model,orientation='vertical',tree_index=1)
```

```python
#Visualizing the confusion matrix
conf_matrix = confusion_matrix(y_test, predictions)         # getting the confusion matrix

classes = ['Addictive disorder', 'Trauma and stress related disorder',
          'Mood disorder', 'Healthy control',
          'Obsessive compulsive disorder', 'Schizophrenia',
          'Anxiety disorder']

#fixing the figure size 
plt.figure(figsize=(10, 10))

# Ploting confusion matrix
plt.imshow(conf_matrix, interpolation='nearest', cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.colorbar()

# Adding labels
tick_marks = np.arange(len(classes))
plt.xticks(tick_marks, classes, rotation=45)
plt.yticks(tick_marks, classes)

# Adding text annotations
thresh = conf_matrix.max() / 2.
for i in range(conf_matrix.shape[0]):
    for j in range(conf_matrix.shape[1]):
        plt.text(j, i, format(conf_matrix[i, j], 'd'),
                 horizontalalignment="center",
                 color="white" if conf_matrix[i, j] > thresh else "black")

plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.tight_layout()
plt.show()
```

```python
#Calculating the cross validation scores 
scores = cross_val_score(model, df, Y_resampled, cv=RepeatedKFold)      
```

```python
print("Cross-validation scores:", scores)
```

## XGBoost

```python
model = xgb.XGBClassifier(device = 'cuda',
                      max_depth=0,
                      eval_metric = 'mlogloss',
                      early_stopping_rounds=10,
                      learning_rate=0.1,
                      n_estimators=100,
                      reg_alpha = 0,
                      reg_lambda = 1,
                    )
```

```python
model.fit(X_train, y_train,
          eval_set=[(X_train,y_train),(X_test, y_test)],
          verbose=False)
```

```python
#visualizing the training process
results = model.evals_result()
plt.plot(results['validation_0']['mlogloss'], label='train')
plt.plot(results['validation_1']['mlogloss'], label='test')
plt.xlabel('Iterations')
plt.ylabel('loss')
plt.title('Training and Validation Loss')
plt.legend()
plt.show()
```
